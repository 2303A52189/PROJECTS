importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset to a apndas dataframe
credit_card_data = pd.read_csv('/content/archive (16).zip',compression='zip')
```

```python
# first 5 rows of the datset
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# info about the dataset
credit_card_data.info()
```

```python
#checking the mssing values in each column
credit_card_data.isnull().sum()
```

```python
# distibution of legit transaction and fraudulent trasaction
credit_card_data['Class'].value_counts()
```

```python
# this dataset is highly unbalenced
# 0--> normal transction
# 1--> fraudulent transaction
# separating the data for the analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# comapre the values for both trasanctions
credit_card_data.groupby('Class').mean()
```

under sampling

buidling a sample datset containing similar distribution of normal transactionn and fradulent transaction

```python
#working on inbalance data
legit_sample = legit.sample(n=492)
```

concatinating two datframes

```python
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

spliting the dataset in to features and tagrget

```python
x=new_dataset.drop(columns='Class',axis=1)
y=new_dataset['Class']
```

training and testing data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

model training

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

model evalution

```python
# arrucary score
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
print('accuracy on training data:',training_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
print('accuracy on test data:',test_data_accuracy)
```

```python
import joblib

# Save the trained model to a file
joblib.dump(model, 'fraud_model.pkl')
```
